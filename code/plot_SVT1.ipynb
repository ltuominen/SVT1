```python
import matplotlib.pyplot as plt
from nilearn import plotting
import os
```

```python
base_path = os.path.dirname(os.getcwd())

svt1 = f'{base_path}/test_subject/mrtm2_sm6/bp.nii.gz'

fig, ax = plt.subplots(figsize=(10, 5))

plotting.plot_stat_map(
    svt1,
    vmin=1,
    vmax=6,
    draw_cross=False,
    black_bg=False,
    axes=ax,

);
fig.suptitle(r'$[^{18}F]\mathrm{SynVesT1}$', fontsize=16)


cbar = fig.axes[-1]  # typically the colorbar is the last axis
cbar.set_ylabel(r'$BP_{ND}$', rotation=90, labelpad=-55, size=14);

fig.savefig(f'{base_path}/figures/test_subject_bpnd_mni152.png')
```
